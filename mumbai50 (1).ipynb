```python
import requests
import pandas as pd
import time

import requests
import pandas as pd
import time

# Replace with your GitHub token
GITHUB_TOKEN = '-'
headers = {'Authorization': f'token {GITHUB_TOKEN}'}
```

```python

# Function to clean up company name
def clean_company_name(company):
    if company:
        company = company.strip()
        if company.startswith('@'):
            company = company[1:]
        return company.upper()
    return ""

# List to store user data
users_data = []

# Step 1: Retrieve users in Mumbai with over 50 followers
def get_users_in_mumbai():
    url = 'https://api.github.com/search/users'
    params = {
        'q': 'location:mumbai followers:>50',
        'per_page': 100
    }
    page = 1

    while True:
        print(f"Fetching page {page}")
        response = requests.get(url, headers=headers, params={**params, 'page': page})
        if response.status_code != 200:
            print(f"Failed to fetch users: {response.json()}")
            break

        users = response.json().get('items', [])
        if not users:
            break

        for user in users:
            username = user['login']
            # Fetch each user's detailed info
            user_details_url = f"https://api.github.com/users/{username}"
            user_response = requests.get(user_details_url, headers=headers)

            if user_response.status_code == 200:
                user_data = user_response.json()
                # Collect the required fields and clean company name
                users_data.append({
                    'login': user_data.get('login', ''),
                    'name': user_data.get('name', ''),
                    'company': clean_company_name(user_data.get('company')),
                    'location': user_data.get('location', ''),
                    'email': user_data.get('email', ''),
                    'hireable': user_data.get('hireable', ''),
                    'bio': user_data.get('bio', ''),
                    'public_repos': user_data.get('public_repos', 0),
                    'followers': user_data.get('followers', 0),
                    'following': user_data.get('following', 0),
                    'created_at': user_data.get('created_at', '')
                })
            time.sleep(1)  # Pause to avoid rate limits

        page += 1
        time.sleep(2)  # Pause between pages to avoid rate limits

```

```python
# Run the function
get_users_in_mumbai()

```

```python
# Step 2: Store data in users.csv
users_df = pd.DataFrame(users_data)
users_df.to_csv('users.csv', index=False)

print("User data saved to users.csv")
```

```python
import pandas as pd

# Load repositories.csv
repos_df = pd.read_csv('/content/repositories.csv')

# Filter out rows with missing languages and missing stargazers_count
repos_df = repos_df[repos_df['language'].notna() & repos_df['stargazers_count'].notna()]

# Calculate the average number of stars for each language
average_stars_per_language = repos_df.groupby('language')['stargazers_count'].mean()

# Find the language with the highest average stars
language_with_highest_avg_stars = average_stars_per_language.idxmax()
highest_avg_stars = average_stars_per_language.max()

# Output the result
print(f"{language_with_highest_avg_stars} has the highest average stars per repository with {highest_avg_stars:.2f} stars.")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load users.csv
users_df = pd.read_csv('/content/users.csv')

# Filter out users without a bio
users_with_bio = users_df[users_df['bio'].notna()]

# Calculate the length of the bio in words
users_with_bio['bio_word_count'] = users_with_bio['bio'].apply(lambda x: len(x.split()))

# Prepare the data for regression
X = users_with_bio[['bio_word_count']]  # Independent variable: bio word count
y = users_with_bio['followers']          # Dependent variable: followers

# Perform linear regression
model = LinearRegression()
model.fit(X, y)

# Get the slope (coefficient) of the regression
slope = model.coef_[0]

# Output the result to 3 decimal places
print(f"{slope:.3f}")
```

```python
import pandas as pd

# Load users.csv
users_df = pd.read_csv('/content/users.csv')

# Calculate the total number of hireable and non-hireable users
total_hireable = users_df[users_df['hireable'] == True].shape[0]
total_non_hireable = users_df[users_df['hireable'] != True].shape[0]

# Calculate the number of users with email for hireable and non-hireable users
email_hireable = users_df[users_df['hireable'] == True]['email'].notna().sum()
email_non_hireable = users_df[users_df['hireable'] != True]['email'].notna().sum()

# Calculate the fractions
fraction_hireable = email_hireable / total_hireable if total_hireable > 0 else 0
fraction_non_hireable = email_non_hireable / total_non_hireable if total_non_hireable > 0 else 0

# Calculate the difference
difference = fraction_hireable - fraction_non_hireable

# Output the result to 3 decimal places
print(f"{difference:.3f}")
```

```python
import pandas as pd
from collections import Counter

# Load users.csv
users_df = pd.read_csv('/content/users.csv')

# Extract surnames (last word of the name)
surnames = users_df['name'].dropna().apply(lambda x: x.strip().split()[-1])

# Count occurrences of each surname
surname_counts = Counter(surnames)

# Find the maximum count
max_count = max(surname_counts.values())

# Get the most common surnames
most_common_surnames = [surname for surname, count in surname_counts.items() if count == max_count]

# Sort surnames alphabetically
most_common_surnames.sort()

# Output the result as a comma-separated string
print(", ".join(most_common_surnames))
```
